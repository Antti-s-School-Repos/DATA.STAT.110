A research on sleep quality aimed to find out how a new medication affects
the sleep quality of people with sleep disorders. Persons who took the part of
research were asked their own opinion on their sleep quality both before the
start of the medication, the variable Y1 = start, and two months after the start
of the medication, the variable Y2 = end. The self-assessment was measured by
the following sleep quality score: 0 = Very poor, 10 = Very good.

The following data was obtained from the subjects regarding the questions:

```python
start = [1, 4, 3, 2, 3, 2, 5, 2, 5, 3, 5, 4]
end = [4, 5, 7, 5, 4, 2, 4, 7, 6, 4, 8, 3]
```

(a) Create a new variable
difference = end − start = Y2 − Y1.
Then calculate the sample mean

```python
difference = [e - s for e, s in zip(end, start)]
difference_mean = sum(difference) / len(difference)
print(difference_mean)
```

b) Calculate the sample variance s^2_y of the variable Y = difference.

```python
sample_variance = sum((x - difference_mean) ** 2 for x in difference) / (len(difference) - 1)
print(sample_variance)
```

c) Calculate the sample median of the variable Y = difference.

```python
import statistics

sample_median = statistics.median(difference)
print(sample_median)
```

d) Classify the values of the Y = difference variable into class intervals:
(−∞, 0], (0, ∞),
where −∞ means minus infinite and ∞ means infinite. 

What is the relative frequency (percentage) of the class interval (0, ∞), i.e., what percentage of
the observed values of the variable Y = difference is between (0, ∞)?

```python
values_above_zero = [x for x in difference if x > 0]
relative_frequency = sum(values_above_zero) / len(difference) * 100
print(relative_frequency)
```
