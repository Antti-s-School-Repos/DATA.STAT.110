a) Create SPSS dataset from the above information. The dataset should con-
tain variables named as height,weight, and gender. Let us denote the variable
height as Y = height. Calculate what is the sum of the values of the variable
Y = height in the dataset, i.e., calculate what is the sum

```python
import numpy as np

males = np.array([(175, 80), (180, 84), (165, 74), (190, 95), (174, 78), (191, 88)], 
                 dtype=[('height', 'i4'), ('weight', 'i4')])

females = np.array([(170, 70), (162, 62), (165, 59), (160, 55), (158, 52), (174, 66)],
                    dtype=[('height', 'i4'), ('weight', 'i4')])

sum_heights = np.sum(males['height']) + np.sum(females['height'])
print(sum_heights)
```

b) What is the sample size n in this study, i.e., how many sampling units i
there are in the dataset?

```python
len(males) + len(females)
```

c) Calculate the sample median of the variable Y = height.

NOTE: Median = Jonon keskimmäinen arvo (jos parillinen määrä niiden keskiarvo)

```python
import statistics

heights = np.concatenate((males['height'], females['height']))
median_height = statistics.median(heights)
print(median_height)
```

d) Create a new body mass index variable
bmi = weight / ((height/100) ∗ (height/100))
Let us denote from now on the variable bmi as Y = bmi.

```python
def calculate_bmi(height, weight):
    return weight / ((height/100) * (height/100))

bmis = []
for person in np.concatenate((males, females)):
    bmi = calculate_bmi(person['height'], person['weight'])
    bmis.append(bmi)

bmi_sample_mean = np.mean(bmis)
print(bmi_sample_mean)

```

e) Create a new variable W = difference, where the observed value wi of the
variable W is calculated as
wi = y_i − ¯y,
when y_i is the observed value of the variable Y = bmi for the sampling
unit i, and ¯y is the sample mean of the variable Y = bmi. After creating
the new variable, calculate the sample standard deviation of the variable
W = difference denoted as sw. In your answer, return only the value of the
sample standard deviation sw.

```python
wi = [bmi - bmi_sample_mean for bmi in bmis]
sw = np.std(wi, ddof=1)
print(sw)
```
